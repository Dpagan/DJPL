Cardio Dataset

Data Features:

Cardio Data is for customers of the treadmill product(s) of a retail store called Cardio Good Fitness. It contains the following variables

Product - treadmill model number

Age - number of years of customers

Gender - Male, Female (gender of customers)

Education - number of years of education that customer has

Marital Status - If the customers are single, married or separated.

Usage - Average times the customer wants to use the treadmill per week

Fitness - Self rated fitness score of the customers (5 - very fit, 1 - very unfit)

Income - Customers income

Miles- expectation of miles to run
